```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['actual'] = iris.target
df.head()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(iris.data)
```

```python
print(f"Inertia: {kmeans.inertia_:.2f}\n")
print(pd.crosstab(df['actual'], df['cluster'],
                  rownames=['Actual'], colnames=['Cluster']))
```

```python
palette = {0: 'red', 1: 'green', 2: 'blue'}

sns.scatterplot(data=df, x=iris.feature_names[0], y=iris.feature_names[1],
                hue='cluster', palette=palette, edgecolor='k', linewidth=0.3)
plt.scatter(*kmeans.cluster_centers_[:, :2].T, marker='X', s=200,
            c=list(palette.values()), edgecolors='k', zorder=5, label='Centroids')
plt.title('KMeans Clustering — Iris (k=3)')
plt.legend()
plt.tight_layout()
plt.show()
```
